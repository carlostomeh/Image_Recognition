```python
import os
import numpy as np
import random
import pandas as pd
import cv2
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import InputLayer,Dense, Conv2D, MaxPool2D, Flatten, RandomFlip, RandomRotation, RandomTranslation 
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import VGG16


from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

### Functions

```python
def extract_class(df):
    list_class   = []
    classes_temp = df.image_pathname.str.split('/')
    
    for i in df.index:
        class_str = classes_temp[i][1]
        list_class.append(class_str)
    df['target'] = list_class
    return df;
```

```python
def preprocess_animals_dataset(df, label_encoder: LabelEncoder, new_dims=(100, 100)):
    # load the images as a feature matrix
    image_list = [] 
    
    for index, img_path in enumerate(df['image_pathname']):
        # read image
        img = cv2.imread('./pubfig83/'+img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
        # image resizing
        # for gray or color images, the linear interpolation sounds good
        img = cv2.resize(img, new_dims, interpolation=cv2.INTER_LINEAR)        
        image_list.append(img)

    # verbose - print every 1000 iterations
    if index % 1000 == 0:
        print(f'{index + 1}/{dataset_df.shape[0]} - {img_path}')
    
    X = np.array(image_list)
    
    # feature scaling
    X = X / 255.0
    
    # encoding the classes
    y = label_encoder.fit_transform(df['target'])
    
    return X, y
```

```python
def plot_history_epochs(history):
    
    history_df = pd.DataFrame(history.history)

    history_df[['loss', 'val_loss']].plot(figsize=(8, 5))
    plt.grid(True)
    # plt.xticks(range(100))
    plt.xlabel('Epochs')
    plt.ylabel('Score')

    history_df[['accuracy', 'val_accuracy']].plot(figsize=(8, 5))
    plt.grid(True)
    # plt.xticks(range(100))
    plt.xlabel('Epochs')
    plt.ylabel('Score')
    
```

### Modeling

##### Basic CNN - Baseline

```python
# Model Function:

def build_cnn_baseline(input_shape, n_classes):
    model = Sequential([
        
        # Basic CNN
        Conv2D(filters=32, kernel_size=(4,4), input_shape=input_shape, activation='relu'),
        MaxPool2D(pool_size=(2,2)),
        Conv2D(filters=32, kernel_size=(4,4), activation='relu'),
        MaxPool2D(pool_size=(2,2)),

        Flatten(),

        Dense(256, activation='relu'),
        Dense(n_classes, activation='softmax')
    ])
    
    return model
```

```python
# parameters

n_classes = len(set(df_train['target']))
input_shape = (64, 64, 3)

# Build model
model_0 = build_cnn_baseline(input_shape, n_classes)

# Build optimizer
opt = tf.keras.optimizers.SGD(learning_rate=0.01)
```

```python
# Compile parameters and optimizer
model_0.compile(loss='sparse_categorical_crossentropy', 
              optimizer=opt, 
              metrics=['accuracy']
             )
```

```python
model_0.summary()
```

##### VGG16 Simple (Fully conected)

```python
base_model = VGG16(include_top=None, 
                   weights="imagenet", 
                   input_shape=(64, 64, 3)) 
base_model.trainable = False
```

```python
base_model.summary()
```

```python
# Model Function:
def build_simple_VGG16(input_shape, n_classes):
    
    model = Sequential([
        base_model,

        Flatten(),

        # FC classifier
        Dense(256, activation='relu'),
        Dense(128, activation='relu'),
        Dense(n_classes, activation='softmax')
    ])
    
    return model
```

```python
early_stopping_cb = tf.keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)


history = model_1.fit(X_train, y_train, 
                    epochs=10, 
                    batch_size=256, 
                    validation_data=(X_val, y_val), 
                    callbacks=[early_stopping_cb])
```

##### VGG16 with Data Augmentation

```python
# Model Function:

def build_VGG16_data_aug(input_shape, n_classes):
    
    model = Sequential([
        InputLayer(input_shape=input_shape),
        
        # data augmentation layers
        RandomFlip("horizontal"),
        RandomRotation(factor=0.1),
        RandomTranslation(height_factor=0.1, width_factor=0.1),
        
        base_model,

        Flatten(),

        # FC classifier
        Dense(256, activation='relu'),
        Dense(128, activation='relu'),
        Dense(n_classes, activation='softmax')
    ])
    
    return model
```

```python
# Parameters
n_classes = len(set(df_train['target']))
input_shape = (64, 64, 3)

# Build model
model_2 = build_VGG16_data_aug(input_shape, n_classes)

# Build optimizer
opt = Adam(learning_rate=0.001)

# Compile parameters and optimizer
model_2.compile(loss='sparse_categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
```

```python
early_stopping_cb = tf.keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)


history_2 = model_2.fit(X_train, y_train, 
                    epochs=10, 
                    batch_size=256, 
                    validation_data=(X_val, y_val), 
                    callbacks=[early_stopping_cb])
```

```python
model_2.evaluate(X_test, y_test)
```

```python
plot_history_epochs(history_2)
```

```python
class_report_results(model_2)
```
